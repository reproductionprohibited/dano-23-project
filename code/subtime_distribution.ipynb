```python
import polars as pl
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import numpy as np
import statsmodels.api as sm
```

```python
df = pd.read_pickle(r"../base_processed.pkl")
```

```python
sns.set_style("darkgrid", rc={'figure.facecolor': 'F9F9F9', 'xtick.color':'red'})
fig, axs = plt.subplots(1, 2, figsize = (16, 7.5), dpi = 600)
sns.histplot(df, x = "sub_time_in_days", color="#6A76D9" , ax = axs[0], bins=50)
sns.histplot(df, x = "sub_time_in_days", color="#6A76D9" , ax = axs[1], log=True, bins=50, alpha = 0.4)
axs[0].set_xticks(np.arange(0,300,30))
axs[0].set_xticklabels(np.arange(0,300,30), color="black")
axs[1].set_xticks(np.arange(0,300,30))
axs[1].set_xticklabels(np.arange(0,300,30), color="black")
axs[0].tick_params(axis='x', which='both', bottom=True)
axs[1].tick_params(axis='x', which='both', bottom=True)
axs[0].set_xlabel(None)
axs[1].set_xlabel(None)
axs[0].set_ylabel(None)
axs[1].set_ylabel(None)
fig.savefig("../images/subtime_distribution.png")
```
