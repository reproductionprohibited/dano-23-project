```python
import polars as pl
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import numpy as np
import statsmodels.api as sm
```

```python
df = pd.read_pickle(r"../base_processed.pkl")
```

```python
new_column = ['0-13','14-17','18-22','23-25','26-30','31-35','36-40','41-45','46-50','51-55','56-60','61-65','65+']
plt.figure(dpi=800)
sns.set_style("darkgrid", rc={'figure.facecolor': 'F9F9F9'})
cp = sns.countplot(df, x="age_category", color="#6A76D9")
cp.set(xlabel = None, ylabel = None, xticklabels=new_column)
plt.xticks(rotation=90)
plt.savefig("../images/age_segment_countplot.png")
```
