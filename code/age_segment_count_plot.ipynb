```python
import polars as pl
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import numpy as np
import statsmodels.api as sm
```

```python
df = pd.read_pickle(r"../base_processed.pkl")
```

```python
new_column = {'0-13':1,
            '14-17':2,
            '18-22':3,
            '23-25':4,
            '26-30':5,
            '31-35':6,
            '36-40':7,
            '41-45':8,
            '46-50':9,
            '51-55':10,
            '56-60':11,
            '61-65':12,
            '65+':13}
plt.figure(dpi=800)
sns.set_style("darkgrid", rc={'figure.facecolor': 'F9F9F9'})
cp = sns.countplot(df, x="age_category", color="#6A76D9")
cp.set(xlabel = None, ylabel = None, xticklabels=list(new_column.keys()))
plt.xticks(rotation=90)
plt.savefig("../images/age_segment_countplot.png")
```
