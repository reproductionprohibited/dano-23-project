```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("data.xlsx")
len(df)
```

```python
df = df.loc[df['age_segment'] != 'Пусто']
```

```python
fig, ax = plt.subplots(figsize=(10, 30))
sns.boxplot(
    df[
        [
            "in_total_1ms_amt",
            "in_total_3ms_amt",
            "in_total_5ms_amt",
            "out_total_1ms_amt",
            "out_total_3ms_amt",
            "out_total_5ms_amt",
        ]
    ]
)
```

```python
len(df[(df['in_total_1ms_amt'] >= df['in_total_1ms_amt'].quantile(0.95))])
```

```python
len(df[(df['in_total_1ms_amt'] >= df['in_total_1ms_amt'].quantile(0.95)) & (df['reject_flg'] == 0)])
```

```python
df['in_total_1ms_amt'].quantile(0.95)
```

```python
all_rich = df[(df['in_total_5ms_amt'] >= df['in_total_5ms_amt'].quantile(0.95))]
len(all_rich)
```

```python
premium_rich = df[(df['in_total_1ms_amt'] >= df['in_total_1ms_amt'].quantile(0.95)) & (df['reject_flg'] == 0)]
len(premium_rich)
```

```python
sns.histplot(data=premium_rich, x='invest_flg')
plt.title("Люди с большим доходом с подпиской")
```

```python
len(df[(df['in_total_1ms_amt'] <= df['in_total_1ms_amt'].quantile(0.15)) & (df['reject_flg'] == 0)])
```

```python
poor_premium = df[(df['in_total_1ms_amt'] <= df['in_total_1ms_amt'].quantile(0.15)) & (df['reject_flg'] == 0)]
```

```python
sns.histplot(data=poor_premium, x='invest_flg')
plt.title("Люди с маленьким доходом с подпиской")
```

```python
non_premium_rich = df[(df['in_total_1ms_amt'] >= df['in_total_1ms_amt'].quantile(0.95)) & (df['reject_flg'] == 1)]

len(non_premium_rich)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=non_premium_rich, x='age_segment', binwidth=20)
plt.title("Люди с высоким доходом без подписки")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=premium_rich, x='age_segment', bins=20)
plt.title("Люди с высоким доходом с подпиской")
```

```python
plt.figure(figsize=(10, 6))
ax = sns.histplot(data=non_premium_rich, x='region_size', binwidth=20)
plt.title("Люди с высоким доходом без подписки")
```
